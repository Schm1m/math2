```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

# 1. Formatieren der Grafik


```python
x_min, x_max, y_min, y_max = -5, 5, -5, 5
fig, ax = plt.subplots(figsize=((x_max - x_min), (y_max - y_min)))
ax.set_aspect("equal")
ax.axis([x_min, x_max, y_min, y_max])
x_ticks = np.arange(x_min, x_max + 1, 1)
y_ticks = np.arange(y_min, y_max + 1, 1)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))

ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))

ax.set_xticks(np.arange(x_min, x_max + 1), minor=True)
ax.set_yticks(np.arange(y_min, y_max + 1), minor=True)
ax.grid(which="both")
ax.minorticks_on()
ax.grid(True)
ax.grid(which="minor", alpha=0.1)
ax.grid(which="major", alpha=0.5)
```

# 2. Berechnen der Spirale und Darstellung

```python
def func(x):
    return np.sqrt(x) * np.array([np.cos(x), np.sin(x)]) # c(t) aber t mit x ersetzt


theta = np.linspace(0, 8* np.pi, 1000)

x_values, y_values = func(theta) # get x and y values from func ( retuns a tuple, which is split to x and y )

ax.plot(x_values, y_values, color="black")

display(fig)
```
