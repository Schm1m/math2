```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,6,num=100)

# 1/4 * ( x^4 - 12x^3 + 49x^2 - 78x + 40 )

fx = []
for i in range(len(x)):
    fx.append((1/4)*(x[i]**4 - 12*x[i]**3 + 49*x[i]**2 - 78*x[i] + 40))

x_min, x_max, y_min, y_max = 0, 6, -1, 2
#fig, ax = plt.subplots(figsize=((x_max-x_min),(y_max-y_min)))
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect('equal')
ax.axis([x_min ,x_max, y_min, y_max])


#x_ticks = np.arange(x_min,x_max+1,2)
x_ticks = [0,1,2,3,4,5,6]
print('x_ticks = ',x_ticks)
ax.set_xticks(x_ticks)
#y_ticks = np.arange(y_min,y_max+1,2)
y_ticks = [-1,0,1]
print('y_ticks = ',y_ticks)
ax.set_xticks(np.arange(x_min,x_max+1),minor=True)
ax.set_yticks(np.arange(y_min,y_max+1),minor=True)
ax.grid(which='both')
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)
markers_x = [1,2,3,4,5]
markers_y = [0,0,1,0,0]
plt.plot(x,fx,'b')
plt.scatter(markers_x, markers_y, color='red')

#display(fig)
```
