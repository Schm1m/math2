# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

# Formatierung

```python
x_min, x_max, y_min, y_max = -3, 3, -1, 5 # dimension of the grid
fig, ax = plt.subplots(figsize=((x_max - x_min), (y_max - y_min)))
ax.set_aspect("equal")
ax.axis([x_min, x_max, y_min, y_max])
x_ticks = np.arange(x_min, x_max + 1, 1)
y_ticks = np.arange(y_min, y_max + 1, 1)

ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))

ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))

ax.set_xticks(np.arange(x_min, x_max + 1), minor=True)
ax.set_yticks(np.arange(y_min, y_max + 1), minor=True)
ax.grid(which="both")
ax.minorticks_on()
ax.grid(True)
ax.grid(which="minor", alpha=0.1)
ax.grid(which="major", alpha=0.5)
```

# Darstellung

```python
l_x = [0,0,1,2,0,2,0,2,2] # x-koordinaten der Punkte von A
l_y = [0,2,4,2,0,0,2,2,0] # y-koordinaten der Punkte von A

ax.plot(l_x, l_y, 'k')

A = [l_x, l_y] # A als Vektor ( für matrix-multiplikation )
deg20 = np.deg2rad(55) # 20° in Bogenmaß
R = [[np.cos(deg20), -np.sin(deg20)], [np.sin(deg20), np.cos(deg20)]] # Transformationsmatrix mit 20° Drehung
T = np.dot(R, A) # matrix-multiplikation
ax.plot(T[0, :], T[1, :], "r")

display(fig)
```
